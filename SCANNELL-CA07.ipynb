```python
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
from nltk.corpus import stopwords
```

```python
finder = BigramCollocationFinder.from_documents(new_sentences)
```

```python
# above finds bigrams and sorts their values using methods below
```

```python
bigram_measures = BigramAssocMeasures() # gets the associative measures of bigrams.
score = bigram_measures.pmi # scores the bigrams using pointwise mutual information
collocations = finder.score_ngrams(score) # this gets the collocations
```

```python
scores = sorted((e[1] for e in collocations)) # sorts their scores
```

```python
median = scores[len(scores)//2] # the median score printed to see what it is
```

```python
# puts the results where its less than the median
results = [each for each in collocations if each[1] >= median]
```
